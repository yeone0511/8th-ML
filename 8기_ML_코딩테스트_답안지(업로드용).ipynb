````python
#ì˜ˆì‹œë‹µì•ˆ
def fibonacci_recursive(n):
    if n == 1 or n == 2:
        return 1
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)
````

````python
### Testing
# ë¬¸ì œì—ì„œ ì œì‹œí•œ 3ê°€ì§€ ì¼€ì´ìŠ¤ì— ëŒ€í•´ í…ŒìŠ¤íŠ¸ í•´ì£¼ì„¸ìš”!

print(fibonacci_recursive(5))  # ì¶œë ¥: 5
print(fibonacci_recursive(7))  # ì¶œë ¥: 13
print(fibonacci_recursive(2))     # ì¶œë ¥: 1
````

### **3-3-2 (6ì )**  
ì£¼ì–´ì§„ ìˆ«ì nì— ëŒ€í•´, në²ˆì§¸ í”¼ë³´ë‚˜ì¹˜ ìˆ˜ì—´ì˜ ê°’ f(n)ì„ ë°˜í™˜í•˜ëŠ” `fibonacci_iterative` í•¨ìˆ˜ë¥¼ êµ¬í˜„í•˜ì„¸ìš”.

**ğŸ“Œì¡°ê±´**
-  **ë°˜ë³µë¬¸** í™œìš©
- ì…ë ¥ `n`ì€ 1 ì´ìƒì˜ ì •ìˆ˜

````python
#ì˜ˆì‹œë‹µì•ˆ
def fibonacci_iterative(n):
    if n == 1 or n == 2:
        return 1

    a, b = 1, 1
    for _ in range(3, n + 1):
        a, b = b, a + b
    return b
````

````python
### Testing
# ë¬¸ì œì—ì„œ ì œì‹œí•œ 3ê°€ì§€ ì¼€ì´ìŠ¤ì— ëŒ€í•´ í…ŒìŠ¤íŠ¸ í•´ì£¼ì„¸ìš”!

print(fibonacci_iterative(5))  # ì¶œë ¥: 5
print(fibonacci_iterative(7))  # ì¶œë ¥: 13
print(fibonacci_iterative(2))  # ì¶œë ¥: 1
````
